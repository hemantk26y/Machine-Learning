```python
try:
    import numpy as np
    import matplotlib.pyplot as plt
    print("Imported Successfully ")
except:
    print("Something went wrong")
```

```python
c1 = np.array(np.random.uniform(2,4, size = (100,2)))
c2 = np.array(np.random.uniform(0,2, size = (100,2)))
plt.scatter([i[0] for i in c1],[i[1] for i in c1])
plt.scatter([i[0] for i in c2],[i[1] for i in c2])
```

```python
cc_c1 = np.array([np.mean([i[0] for i in c1]),np.mean([i[1] for i in c1])])
cc_c2 = np.array([np.mean([i[0] for i in c2]),np.mean([i[1] for i in c2])])
plt.scatter([i[0] for i in c1],[i[1] for i in c1])
plt.scatter([i[0] for i in c2],[i[1] for i in c2])
plt.scatter(cc_c1[0],cc_c1[1])
plt.scatter(cc_c2[0],cc_c2[1])
```

```python
x = int(input("enter the x coordinate :"))
y = int(input("enter the y coordinate :"))
t = np.array([x,y])
t
```

```python
dis1 = np.linalg.norm(t - cc_c1)
dis2 = np.linalg.norm(t - cc_c2)
plt.scatter([i[0] for i in c1],[i[1] for i in c1])
plt.scatter([i[0] for i in c2],[i[1] for i in c2])
plt.scatter(cc_c1[0],cc_c1[1])
plt.scatter(cc_c2[0],cc_c2[1])
plt.scatter(x,y)
if dis1<dis2:
    print("point belong to c1")
else:
    print("point belong to c2")
```
