# K-Means on IRIS DataSet With Sklearn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# importing data

data_iris = pd.read_csv("iris.csv")
data_iris
```

```python
# making data available only with features 

x_iris = data_iris.drop(['y'],axis = 1)
x_iris
```

```python
# takig mean of all features 

d = x_iris.to_numpy()
m = np.mean(d , axis = 1)
d,m
```

```python
# plotting data

plt.scatter(x_iris['x1'],[i for i in m])
plt.scatter(x_iris['x2'],[i for i in m])
plt.scatter(x_iris['x3'],[i for i in m])
plt.scatter(x_iris['x4'],[i for i in m])
```

```python
# applying k-means through sklearn

km_iris = KMeans(n_clusters= 3).fit(x_iris)
```

```python
# centers for iris dataset

centers_iris = km_iris.cluster_centers_
centers_iris
```

```python
plt.scatter(x_iris['x1'],[i for i in m])
plt.scatter(centers_iris[:,0],centers_iris[:,0],color='black')
plt.scatter(x_iris['x2'],[i for i in m])
plt.scatter(centers_iris[:,1],centers_iris[:,1],color='pink')
plt.scatter(x_iris['x3'],[i for i in m])
plt.scatter(centers_iris[:,2],centers_iris[:,2],color='brown')
plt.scatter(x_iris['x4'],[i for i in m])
plt.scatter(centers_iris[:,3],centers_iris[:,3],color='red')
```

```python
# plotting the graph

plt.scatter(x_iris['x1'],x_iris['x2'])
plt.scatter(centers_iris[:,0],centers_iris[:,1],color='black')
plt.scatter(x_iris['x2'],x_iris['x3'])
plt.scatter(centers_iris[:,2],centers_iris[:,3],color='pink')
```

```python
# labels

labels_iris = km_iris.labels_
labels_iris
```

```python
# prediction through prediction method

y_pred_iris = km_iris.predict(x_iris)
y_pred_iris
```

```python
# Accuracy from scratch

Accuracy_iris = ((sum(labels_iris==y_pred_iris)/len(labels_iris))*100).round(2)
print("Accuracy : ",Accuracy_iris," %")
```
