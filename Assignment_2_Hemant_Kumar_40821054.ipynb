```python
try:
    import numpy as np
    import matplotlib.pyplot as plt
    print("Imported Successfully")
except:
    print("Something went wrong")
```

```python
c1 = np.array(np.random.uniform(2,4, size = (100,1000)))
c2 = np.array(np.random.uniform(0,2, size = (100,1000)))
plt.scatter([i[0] for i in c1],[i[999] for i in c1])
plt.scatter([i[0] for i in c2],[i[999] for i in c2])
```

```python
m_c1 = np.mean(c1, axis= 0)
m_c2 = np.mean(c2, axis=0)
plt.scatter([i[0] for i in c1],[i[1] for i in c1])
plt.scatter([i[0] for i in c2],[i[1] for i in c2])
plt.scatter(m_c1[0],m_c1[999])
plt.scatter(m_c2[0],m_c2[999])
```

```python
x = int(input("enter the x coordinate :"))
y = int(input("enter the y coordinate :"))
t = np.array(np.random.uniform(x,y,[1000]))
plt.scatter([i[0] for i in c1],[i[1] for i in c1])
plt.scatter([i[0] for i in c2],[i[1] for i in c2])
plt.scatter(m_c1[0],m_c1[999])
plt.scatter(m_c2[0],m_c2[999])
plt.scatter(x,y)
dis1 = np.linalg.norm(t-m_c1) 
dis2 = np.linalg.norm(t-m_c2) 
if dis1<dis2:
    print("Point belongs to cluster 1")
else:
    print("Point belongs to cluster 2")
```
